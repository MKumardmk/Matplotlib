```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## `Colored Scatterplots`

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.show()
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'],
           cmap='winter') 

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'],
           cmap='winter')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar() # adding the colour bar
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'],
           cmap='winter', alpha=0.8)
# alpha parameter will show the instensity of the colour

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar()
plt.show()
```

## `Plot size`

```python
# increase or decrease the graph size
# for that use the function plt.figure(figsize=(6,6))

plt.figure(figsize=(10,6))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'],
           cmap='winter', alpha=.8)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar()
plt.show()
```

```python
df = pd.read_csv('batsmans.csv')
df
```

```python
df.shape
```

```python
sample_df = df.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'] )
plt.show()
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'] )

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],
            sample_df['batter'].values[i])

plt.show()
```

```python
# increase the figure size
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'], sample_df['strike_rate'] )

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],
            sample_df['batter'].values[i])

plt.show()
```

```python
'''we want to show the size of the each dot 
based on the runs scored by the batsmen'''

plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'] )
# including the s parameter for showing the size of the data point

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],
            sample_df['batter'].values[i])

plt.show()
```

## `Horizontal and Vertical lines`

```python
'''suppose we want batsmen who has strike rate of 130,
for that we can plot a line'''

plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='black')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],
             sample_df['strike_rate'].values[i],
             sample_df['batter'].values[i])
    
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1)

ax[0].scatter(df['avg'],df['strike_rate'])
ax[1].scatter(df['avg'],df['runs'])

ax[0].set_title("avg Vs strike rate")
ax[0].set_ylabel("strike rate")

ax[1].set_title("avg Vs runs")
ax[1].set_ylabel("runs")
ax[1].set_xlabel("avg")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True, figsize = (10,8))

ax[0].scatter(df['avg'],df['strike_rate'], c="red")
ax[1].scatter(df['avg'],df['runs'], c='green')

ax[0].set_title("avg Vs strike rate")
ax[0].set_ylabel("strike rate")

ax[1].set_title("avg Vs runs")
ax[1].set_ylabel("runs")
ax[1].set_xlabel("avg")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10)) #we will get 2D array axis

ax[0,0].scatter(df['avg'],df['strike_rate'])  
ax[0,0].set_title('avg vs strike_rate')

ax[0,1].scatter(df['avg'],df['runs'])
ax[0,1].set_title('avg vs runs')

ax[1,0].hist(df['avg'])
ax[1,0].set_title('avg')

ax[1,1].hist(df['runs'])
ax[1,1].set_title('runs')

plt.show()
```

>* another way to create subplots

```python
fig = plt.figure(figsize=(9,9))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(df['avg'],df['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(df['runs'])

ax3 = fig.add_subplot(2,2,4)
ax3.hist(df['avg'])

plt.show()
```

## `3D Surface plot`

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
xx.shape
```

```python
# we want to create z with help of x and y
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z)

plt.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)

z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')

fig.colorbar(p)
plt.show()
```

## `Contour plot`

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)

z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap='viridis')

fig.colorbar(p)
plt.show()
```

```python
# filled cointour graph

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')

fig.colorbar(p)
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)

z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')

fig.colorbar(p)
plt.show()
```
